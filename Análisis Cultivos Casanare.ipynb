# Cargar Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("/content/Explotaci_n_Agr_cola_del_Departamento_de_Casanare_20241130.csv")
data
```

# Eliminación de valores nulos en la columna PRODUCCIÓN (Ton)

```python
# Extraer los valores nulos de la producción
valores_nulos = data[data['PRODUCCIÓN (Ton)'].isnull()]
# Mostrar los valores nulos
print("Valores nulos en 'PRODUCCIÓN (Ton)':")
print(valores_nulos)

# Contar cuántos valores nulos hay
cantidad_nulos = valores_nulos.shape[0]
print(f"\nTotal de valores nulos: {cantidad_nulos}")

# Eliminar las filas donde 'Área Sembrada (ha)' tiene valores nulos
data_sin_nulos = data.dropna(subset=['PRODUCCIÓN (Ton)'])

# Mostrar las primeras filas del dataframe resultante
print("Datos después de eliminar los valores nulos en PRODUCCIÓN (Ton)':")
print(data_sin_nulos.head())

# Contar el número de filas restantes
print(f"\nTotal de filas después de eliminar valores nulos: {data_sin_nulos.shape[0]}")
```

# Normalizamos los nombres de los cultivos.

```python
import pandas as pd
from unidecode import unidecode
data_sin_nulos['CULTIVO'] = data_sin_nulos['CULTIVO'].astype(str)
data_sin_nulos['Cultivo_normalizado'] = data_sin_nulos['CULTIVO'].str.lower().apply(unidecode).str.strip()
data_sin_nulos['Cultivo_normalizado']
```

# Análisis de los 5 cultivos con mayor producción

```python
# Asegurarse de que la columna "Producción" sea numérica
data_sin_nulos['PRODUCCIÓN (Ton)'] = pd.to_numeric(data_sin_nulos['PRODUCCIÓN (Ton)'], errors='coerce')

# Agrupar por "CULTIVO" y sumar la producción
produccion_por_cultivo = data_sin_nulos.groupby('Cultivo_normalizado')['PRODUCCIÓN (Ton)'].sum().reset_index()

# Ordenar por producción en orden descendente
produccion_por_cultivo = produccion_por_cultivo.sort_values(by='PRODUCCIÓN (Ton)', ascending=False)

# Seleccionar los 5 cultivos con mayor producción
top_5_cultivos = produccion_por_cultivo.head(5)

# Mostrar los resultados
print("Los 5 cultivos con mayor producción:")
print(top_5_cultivos)
```

# Limpieza de datos: Eliminar o imputar valores faltantes.
Convertir variables categóricas como Cultivo en etiquetas numéricas.



```python
from sklearn.preprocessing import LabelEncoder
data_sin_nulos['CULTIVO'] = LabelEncoder().fit_transform(data_sin_nulos['CULTIVO'])
data_sin_nulos['CULTIVO']
```

# Calcular la moda, la mediana y la desviación estándar para "Área Sembrada (ha) y escalamiento de los datos en logaritmo con base 10

```python
# Asegurarse de que la columna 'Área Sembrada (ha)' sea numérica
data_sin_nulos['"Área Sembrada (ha)"'] = pd.to_numeric(data_sin_nulos['"Área Sembrada (ha)"'], errors='coerce')

# Agrupar por 'Cultivo_normalizado' y calcular los estadísticos (media, desviación estándar, mediana)
estadisticos_por_cultivo = data_sin_nulos.groupby('Cultivo_normalizado')['"Área Sembrada (ha)"'].agg(['mean', 'std', 'median'])

# Calcular la moda por cultivo, aplicando la función lambda sobre cada grupo
moda_por_cultivo = data_sin_nulos.groupby('Cultivo_normalizado')['"Área Sembrada (ha)"'].apply(lambda x: x.mode()[0])

# Unir la moda con los demás estadísticos
estadisticos_por_cultivo['mode'] = moda_por_cultivo

# Escalamiento logarítmico base 10 de 'Área Sembrada (ha)'
data_sin_nulos['Log10 Área Sembrada (ha)'] = np.log10(data_sin_nulos['"Área Sembrada (ha)"'].replace(0, np.nan))  # Evitar log(0)

# Agrupar nuevamente para calcular la media logarítmica por cultivo
log10_areas_por_cultivo = data_sin_nulos.groupby('Cultivo_normalizado')['Log10 Área Sembrada (ha)'].mean().reset_index()

# Unir los datos de estadísticos con los logaritmos
estadisticos_completos = pd.merge(estadisticos_por_cultivo, log10_areas_por_cultivo, on='Cultivo_normalizado')

# Mostrar los resultados
print("Estadísticos y Logaritmos para cada cultivo:")

indices_especificos = [3, 6, 38, 5, 44]
estadisticos_completos.iloc[indices_especificos]
```

```python
import seaborn as sns
data['Log10 Producción (Ton)'] = np.log10(data_sin_nulos['PRODUCCIÓN (Ton)'].replace(0, np.nan))
data['Log10 Área Cosechada (ha)'] = np.log10(data_sin_nulos['"Área Cosechada (ha)"'].replace(0, np.nan))
data['Log10 Área Sembrada (ha)'] = np.log10(data_sin_nulos['"Área Sembrada (ha)"'].replace(0, np.nan))

# Agrupar por 'Cultivo_normalizado' y calcular la media de las variables normalizadas
log10_areas = data_sin_nulos.groupby('Cultivo_normalizado')[['Log10 Producción (Ton)',
                                                  'Log10 Área Sembrada (ha)',
                                                  'Log10 Área Cosechada (ha)']].mean().reset_index()

# Calcular la matriz de correlación
correlation_matrix = log10_areas[['Log10 Producción (Ton)',
                                  'Log10 Área Sembrada (ha)',
                                  'Log10 Área Cosechada (ha)']].corr()

# Graficar la matriz de correlación como un heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Matriz de Correlación entre Producción y Áreas Normalizadas')
plt.show()
```
